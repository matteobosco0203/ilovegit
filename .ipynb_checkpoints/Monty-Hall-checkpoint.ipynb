```python
newcomer=[999]
switcher=[999]
conservative=[999]
for i in range(1000):
    doors=[0,0,1]
    co=["closed", "closed", "closed"]
    choice=np.random.randint(0,2)
    co[choice]="chosen"

    if doors[choice]==1:
        conservative=np.append(conservative,1)
    else:
        conservative=np.append(conservative,0)

    if doors[choice]==0:
        switcher=np.append(switcher,1)
    else:
        switcher=np.append(switcher,0)
    
    n_choice=np.random.randint(0,1)
    
    if doors[n_choice]==0:
        newcomer=np.append(newcomer,0)
    else:
        newcomer=np.append(newcomer,1)
    
conservative=np.delete(conservative,0)
switcher=np.delete(switcher,0)
newcomer=np.delete(newcomer,0)

print(conservative, switcher, newcomer)
```
